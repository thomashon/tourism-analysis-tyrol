```python
cities = df_2.City.unique()
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="specify_your_app_name_here")
```

```python
import time
```

```python
latitudes = []
longitudes = []

for city in cities:
    print(city)
    location = geolocator.geocode(city)
    latitudes.append(location.latitude)
    longitudes.append(location.longitude)
    time.sleep(0.3)

```

```python
lat_long = pd.DataFrame()
```

```python
lat_long['City'] = cities
lat_long['Latitude'] = latitudes
lat_long['Longitude'] = longitudes
```

```python
df_2 = pd.merge(df_2, lat_long, on='City', how='inner')
```
