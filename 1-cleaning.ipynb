```python
import pandas as pd
import numpy as np
import glob
```

```python
paths = glob.glob('data/xls/ue-mon*.xls')
```

```python
paths
```

```python
for i in range(len(paths)):
    df = pd.read_excel(paths[i])
    
    df.columns = ['Jahr', 'Bez', 'Gemnr', 'Gemeinde', 'Jänner', 'Februar', 'März',
       'April', 'Mai', 'Juni', 'Juli', 'August', 'September', 'Oktober',
       'November', 'Dezember']
    
    df.Gemeinde = df.Gemeinde.str.strip()
    
    monate = df[df.columns[4:]]
    unstacked = monate.unstack()
    nights = list(unstacked)
    
    new_nights = []
    for j in range(279):
        for a in range(j, 3348, 279):
            new_nights.append(nights[a])
    
    years = [2000 + i] * len(nights)
    gemeinden = []
    
    for i in range(len(df.Gemeinde.unique())):
        for j in range(12):
            gemeinden.append(df.Gemeinde.unique()[i])
            
    months = []
    for j in range(279):
        for i in range(1,13):
            months.append(i)
            
    test = {
    'year': years,
    'city': gemeinden,
    'month': months,
    'overnight_stays': new_nights
    }
    
    test_df = pd.DataFrame(test)
    
    gezipped = zip(df.Gemeinde, df.Bez)
    bezirks_mapper = dict(gezipped)
    test_df['bezirk'] = test_df.city.map(bezirks_mapper)
    
    gezipped = zip(df.Gemeinde, df.Gemnr)
    gemnr_mapper = dict(gezipped)
    test_df['nr'] = test_df.city.map(gemnr_mapper)
    
    test_df.to_csv('data/csv/' + str(years[0]) + '.csv')
```
